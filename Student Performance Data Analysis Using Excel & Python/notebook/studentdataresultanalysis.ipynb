```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
v = pd.read_csv("studentInfo.csv")

# Clean column names
v.columns = v.columns.str.strip()

# Convert to numeric
v["Age"] = pd.to_numeric(v["Age"], errors="coerce")
v["Marks"] = pd.to_numeric(v["Marks"], errors="coerce")

# Remove invalid rows
v = v.dropna(subset=["Age", "Marks"])

# Calculate average marks by age
avg_marks_age = v.groupby("Age")["Marks"].mean()

# Print result
print(avg_marks_age)
print(type(avg_marks_age))

# Plot
avg_marks_age.plot(kind="bar")
plt.xlabel("Age")
plt.ylabel("Average Marks")
plt.title("Average Marks by Age")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

v = pd.read_csv("studentInfo.csv")

sns.scatterplot(x="Age", y="Marks", data=v)
plt.title("Age vs Marks")
plt.show()

plt.scatter(v["Age"], v["Marks"])
plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()
```

```python
plt.hist(v["Marks"], bins=10)
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()
```

```python
top_10 = v.sort_values(by="Marks", ascending=False).head(10)
print("Top 10 Students:")
print(top_10[["Name", "Roll No", "Marks", "Age"]])
```

```python
correlation = v["Age"].corr(v["Marks"])
print("Correlation between Age and Marks:", correlation)
```

```python
v["Result"] = v["Marks"].apply(lambda x: "Pass" if x >= 40 else "Fail")
print(v[["Name", "Marks", "Result"]].head())
```
